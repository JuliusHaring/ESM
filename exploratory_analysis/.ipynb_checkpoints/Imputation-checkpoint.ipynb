```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_rows = 999
```

```python
df = pd.read_csv("../data/raw.csv", parse_dates=["ESM_START", "ESM_END"])
columns_delete = ['CRF_IBS_subtypecalc', 'ESM_ABDPAIN_LOC','ESM_SUBS','ESM_SUBS_CAF',
 'ESM_SUBS_NIC',
 'ESM_SUBS_ALC',
 'ESM_SUBS_DRUG',
 'ESM_SUBS_MED',
 'ESM_SUBS_NONE',
 'ESM_END']
columns = [col for col in df.columns if ("ESM_" in col or "CRF_" in col or "REGISTRATION_ID" in col or "DAY" == col or "MEASUREMENT_MOMENT" ==  col) and col not in columns_delete and "ESM_MORN" not in col and "ESM_MED" not in col]
df = df[columns]
df.head()
```

# NAN for Pain location

```python
location_columns =  [col for col in df.columns if "ESM_ABDPAIN_" in col]

ilocs= []

for idx, row in enumerate(df[location_columns][df.isna()].values):
    for val in row:
        if(val >= 0):
            ilocs.append(idx)
ilocs
```

If we have NaN in location, the whole row has no information.

```python
#df[df["ESM_ABDPAIN"]==0][location_columns] = 0
idx_abdpain_zero = df[df["ESM_ABDPAIN"]==0].index
col_abdpain_location = []
for idx, col in enumerate(df.columns):
    if "ESM_ABDPAIN_" in col:
        col_abdpain_location.append(idx)
df.iloc[idx_abdpain_zero, col_abdpain_location]= 0
```

If no pain present, locations must be 0 too.
